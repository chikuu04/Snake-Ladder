```python
import random
```

```python
# defined the constant values
max_val = 100
dice_face = 6
```

```python
# snakes from where you go down 
snakes = {
    17 : 7,
    54 : 34, 
    62 : 19, 
    98 : 79
}
# ladders from where you go up
ladders = {
    3 : 38,
    24 : 33,
    42 : 93,
    72 : 84
}
```

```python
# snake found 
snake_founds = [
    "OHH NOOO!!",
    "SNAKE BITE!!",
    "TIME TO GO DOWN HILL!!"
]
# ladder found 
ladder_founds = [
    "OHH YESSS!!!",
    "GOT THE LADDER!!",
    "TIME TO GO UP HILL!!"
    
]
```

```python
# First we print the rules and welcome message to user
def welcome_message():
    message = """
     Rules:
     1. Initially both the players are at 0 position.
     2. The one who reaches 100 first will be the winner.
     3. In your path there will be snakes and ladders
          a) Ladder will make your path shorter.
          b) Snakes will make your path longer.
     4. There are two modes in the game:
          a) Write "roll" for the auto mode.
          b) Write no. between 0 and 20 for manual mode.
     5. To exit the game in between write "quit".
        
     ############# LETS START THE GAME ############## 
     """
    print(message)
```

```python
# Create a function containing the names of two players
def player_names():
    player1 = None
    # we check whether the input is valid or not
    while not player1:
        player1 = input("Enter the name of player 1: ")
    player2 = None
    while not player2:  
        player2 = input("Enter the name of player 2: ")
    print("\nGAME IS IN BETWEEN " + player1 + " AND " + player2 + "\n")
    return player1 , player2
```

```python
# create a function that gives a random value between 1 and 6 , when we roll the dice
def roll():
    val = random.randint(1,dice_face)
    print("Its a "+str(val))
    return val
```

```python
# create a function for snake bites
def snake_bites(prev_val , new_val , player_name):
    print("\n"+ random.choice(snake_founds) + "~~~~~~>")
    print("\n"+ player_name + " goes from " + str(prev_val) + " to "+ str(new_val))
    
    
    
```

```python
#create a function for ladder found
def ladder_found(prev_val , new_val, player_name):
    print("\n"+ random.choice(ladder_founds) + "$$$$$$$$")
    print("\n"+ player_name + " goes from "+ str(prev_val) + " to "+ str(new_val))
```

```python
# create a function to check whether the player found the ladder or snake
def ladder_snake(player_name , current_value , dice_value):
    prev_value = current_value
    current_value = current_value + dice_value
    
    if current_value > max_val:
        return prev_value
    
    print("\n"+ player_name + " moves from "+ str(prev_value) + " to "+str(current_value))
    
    #now check whether the player is in snake or ladder
    
    if current_value in snakes:
        final_value = snakes.get(current_value)
        snake_bites(current_value,final_value,player_name)
        
    elif current_value in ladders:
        final_value = ladders.get(current_value)
        ladder_found(current_value,final_value,player_name)
        
    else:
        final_value = current_value
        
    return final_value    
```

```python
# check for win 
def win(player_name , pos):
    if pos == max_val:
        print("\n" + player_name + " Won the game!!****** Congratulations ****")
        return False
    else:
        return True
```

```python
# function for the game 
def game():
    welcome_message()
    player1,player2 = player_names()
    
    #intial position of both the players
    player1_current_pos = 0
    player2_current_pos = 0
    
    flag = 1
    boolean = True
    
    ###################code for player 1############
    
    while boolean:
        
        if (flag != 0):
            flag = 0
            print(player1 + "'s turn")
            
            input1 = input()
            
            # check whether the input is for auto mode or manual mode
            if input1.isnumeric():
                input1 = int(input1)
                if input1>0 and input1<20:
                    dice_value = input1
                else:
                    print("please enter the value between 0 and 20")
                    flag = 1
                    
            elif input1.isalpha():
                input1 = input1.lower()
                if input1 == "roll":
                    dice_value = roll()
                elif input1 == "quit":
                    player2_current_pos = 100
                    boolean = win(player2,player2_current_pos)
                    flag =1
                else:
                    print("enter the valid input")
                    flag = 1
            else:
                print("enter the valid input")
                flag = 1
                
            if (flag != 1):
                player1_current_pos = ladder_snake(player1, player1_current_pos, dice_value)

                boolean = win(player1, player1_current_pos)
                if boolean == False:
                    break
   ###################code for player 2################

        if (flag != 1):
            flag = 1
            print(player2 + "'s turn")
            input2 = input()
            #check whether the input is for auto mode or manual mode
            if input2.isnumeric():
                input2 = int(input2)
                if input2>0 and input2<20:
                    dice_value = input2
                else:
                    print("please enter the value between 0 and 20")
                    flag = 0
                    
            elif input2.isalpha():
                input2 = input2.lower()
                if input2 == "roll":
                    dice_value = roll()
                elif input2 == "quit":
                    player1_current_pos = 100
                    boolean = win(player1,player1_current_pos) 
                    flag =0
                else:
                    print("enter the valid input")
                    flag = 0
            else:
                print("enter the valid input")
                flag = 0
                
            if (flag != 0):
                player2_current_pos = ladder_snake(player2, player2_current_pos, dice_value)

                boolean = win(player2, player2_current_pos)  
                if boolean == False:
                    break

            
```
